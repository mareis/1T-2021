# Halveringsmetoden
---

<img src='https://upload.wikimedia.org/wikipedia/commons/d/d9/Bisection_anime.gif'>

Halveringsmetoden er ein enkel numerisk metode for å finna et nullpunkt til ein funksjon. Vi finner altså en løsning på likninga $f(x) = 0$. 

For å bruka metoden må vi kjenne to punkter a og b som er slik at $f(a)$ og $f(b)$ har ulikt fortegn, og  funksjonen må være kontinuerlig i intervallet $[a,b]$. Vi startar med å finna midtpunktet m mellom $a$ og $b$. Hvis  $f(m)$ og $f(a)$ har ulikt fortegn, så leter vi videre i intervallet $[a,m]$. Hvis ikke fortsetter vi med intervallet $[m,b]$. Slik kan vi fortsette helt til vi er fornøyd med svaret!

**Algoritme**

1. Regn ut m, midtpunktet til intervallet $[a,b]$, $m = (a + b)/2$.
2. Regn ut funksjonsverdien for midtpunktet, m, $f(m)$.
3. Undersøk fortegnet til verdien av $f(m)$ og erstatt enten $(a, f(a))$ eller $(b, f(b))$ med $(m, f(m))$ slik at nullpunktet eksisterer i det nye intervallet.


**Pseudokode:**

```
 Gitt funksjonen f, a og b.
        gjenta N ganger
        	m = (a + b)/2
        	hvis f(a)・f(x) < 0
        		b = m
        	ellers
        		a = m
    
```

````python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
````

## Implementasjon av halveringsmetoden
---

````python
def halveringsmetoden(f, a, b, N):
    m_list = []; 
    for i in range(N):
        m = (a + b)/2
        if f(a)*f(m) < 0:
            b = m
        else:
            a = m
        
        m_list.append(m);
    return m, m_list
````

````python
 halveringsmetoden(f, -9, 10, 10)
````

## Noen eksempler
---

### Eksempel 1 
---

**Løs likningen, $f(x)=2x-4=0$, numerisk.**


Vi starter med å definere $f(x)=2x-4$

````python
def f(x):
    return 2*x-4
````

Løser likningen $f(x)=0$ ved hjelp av halveringsmetoden

````python
x = halveringsmetoden(f, -10, 10, 17)
````

Skriver ut løsningen til likningen

````python
print(f'x = {x[0]:.4f}')
````

Likningen $f(x)=2x-4=0$ har løsningen:

$$x_1 \approx 2.0000$$

### Eksempel 2
---

**Løs likningen, $f(x)=x^2-16=0$, numerisk.**

Vi starter med å definere $f(x)=x^2-16$

````python
def g(x):
    return x**2 - 16
````

Det er lurt å tegne grafen til funksjonen før vi bestemmer området vi skal lete etter nullpunktene.

````python
x = np.linspace(-6,6,1000) # 1000 x-verdier fra -6 til 6
````

````python
plt.xlabel('$x$') # Merker x-aksen
plt.ylabel('$g(x)$') # Merker y-aksen
plt.plot(x, g(x), label='$g(x)=x^2-16$') # Tegner g(x) 
plt.legend() # Infoboksen øverst i hjørnet. Bruker data fra label
plt.show()
````

Vi ser at $g(x)$ har 2 nullpunkt. 

Det første nullpunktet finner vi mellom -6 og 0

````python
x_1 = halveringsmetoden(g, -6, 0, 18)
print(f'x = {x_1[0]:.4f}')
````

Det andre nullpunktet ligger mellom 0 og 6

````python
x_2 = halveringsmetoden(g, 0, 6, 17)
print(f'x = {x_2[0]:.4f}')
````

Likningen $f(x)=x^2-16=0$ har løsningene:

$$x_1 \approx -4.0000 \wedge x_2 \approx 4.0000$$

Plot som viser tilnærmelsen til nullpunktet for hver iterasjon.

````python
plt.xlabel('$n$') # Merker x-aksen
plt.ylabel('Nullpunkt, $x_1$') # Merker y-aksen
plt.plot(x_1[1], 'r', label='Tilnærmingsverdi for det ene nullpunktet') # Tegner g(x) 
plt.legend() # Infoboksen øverst i hjørnet. Bruker data fra label
plt.show()
````

### Eksempel 3
---

**Løs likningen, $h(x) = e^x + x^2 − x − 4=0$, numerisk.**

Grunnen til at vi løser likninger numerisk er fordi de ikke lar seg løse eksakt.

<img src='CAS.png' width='300px'>

Vi starter med å definere $h(x) = e^x + x^2 − x − 4$

````python
def h(x):
    return np.exp(x) + x**2 - x - 4
````

Det er lurt å tegne grafen til funksjonen før vi bestemmer området vi skal lete etter nullpunktene.

````python
x = np.linspace(-2,2,1000)

plt.xlabel('$x$') # Merker x-aksen
plt.ylabel('$h(x)$') # Merker y-aksen
plt.plot(x, h(x), label='$h(x)= e^x + x^2 − x − 4$') # Tegner g(x) 
plt.legend() # Infoboksen øverst i hjørnet. Bruker data fra label
plt.show()
````

````python
x_1 = halveringsmetoden(h, -2, 0, 17)
print(f'x = {x_1[0]:.4f}')
````

````python
x_2 = halveringsmetoden(h, 0, 2, 17)
print(f'x = {x_2[0]:.4f}')
````

Likningen $h(x) = e^x + x^2 − x − 4=0$ løsningene:

$$x_1 \approx -1.5071 \wedge x_2 \approx 1.2887$$

Plot som viser tilnærmelsen til nullpunktet for hver iterasjon.

````python
plt.xlabel('$n$') # Merker x-aksen
plt.ylabel('Nullpunkt, $x_1$') # Merker y-aksen
plt.plot(x_1[1], 'r', label='Tilnærmingsverdi for nullpunktet') # Tegner g(x) 
plt.legend() # Infoboksen øverst i hjørnet. Bruker data fra label
plt.show()
````

# Newtons metode

````python
from scipy.misc import derivative
def derivertN(f, a, N, tol=1E-13):
    m = a
    i = 0
    m_list = []; 
    while i <= N and np.abs(f(m)) >= tol:
        m = a - f(a)/derivative(f, a, tol)
        a = m
        i += 1
        m_list.append(m);

    return m, m_list
````
