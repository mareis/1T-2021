### ( i )    Utled og forklar Newtons metode.

<img src='newtons.png' width='800px'>

Utgangspungt fra tekste på Wikipedia

Newtons metode, også kjent som Newton-Raphson-metoden, er en metode for å finne nullpunkter til funksjoner. Man finner ikke en eksakt løsning, men en tilnærmelsesverdi med så høy nøyaktighet man ønsker. Metoden går ut på at man starter fra et punkt x_0 i nærheten av et nullpunkt og bruker dette punktet som tilnærmelsesverdi. Så finner man grafens tangent (den grå tangenten på figuren over) i punktet $(x_0, f(x_0)$, og bruker tangentens skjæringspunkt med x-aksen som ny tilnærmelsesverdi x_1. Så finner man grafens tangent (den røde tangenten på figuren over) i punktet $(x_1, f(x_1)$, og bruker tangentens skjæringspunkt med x-aksen som ny tilnærmelsesverdi $x_2$. Prosessen gjentas til man har fått ønsket nøyaktighet. Regneprosessen resulterer i følgende rekursjonsforme


For å komme frem til Newtons metode tar vi utganspunkt i topunktsformelen $y_1-y_0=a(x_1-x_0)$. For å gjøre den mere generell kan vi skrive den slik $y_{n+1}-y_n=a(x_{n+1}-x_n)$ og videre $f(x_{n+1})-f(x_n)=a(x_{n+1}-x_n)$.

Siden vi ønsker å finne ut når tangenten skjærer x-aksen (nullpunktet til tangenten) så setter vi $f(x_{n+1})=0$. 

Med $a=f^\prime(x_n)$ så løser vi med hensyn på $x_{n+1}$

\begin{align}
-f(x_n) & = f(x_{n+1})(x_{n+1}-x_n)\\
-\frac{f(x_n)}{f(x_{n+1})} & = x_{n+1}-x_n\\
x_{n+1} & = x_n-\frac{f(x_n)}{f^\prime(x_n)}\\
\end{align}





###   ( ii )   Gjør oppgave 9.9 fra bok

```python
import numpy as np
theta_verdier = [np.pi/3, np.pi/4, np.pi/5, np.pi/6, np.pi/7, np.pi/8, np.pi/9, np.pi/10]
kast_tider = []
kast_x_lengder = []
kontrolltider = []
```

$x(t)=18cos(\theta)t$

```python
def x(t, theta=np.pi/10):
    return 18*np.cos(theta)*t
```

$y(t)=-\frac12 g t^2 + 18sin(\theta)t$

```python
def y(t, theta=np.pi/10, g=9.81):
    return -(1/2)*g*t**2 + 18*np.sin(theta)*t
```

**Newtons metode for å finne når ballen treffer bakken**

```python
def newtons_metode(f, theta, t=2, N=50, h=10**-5):
    for n in range(N):
        t = t - f(t, theta)/((f(t + h, theta) - f(t, theta))/h)    
    return t
```

**Kjører newtons metode for alle de aktuelle $\theta$- verdiene**

```python
for theta in theta_verdier:
    kast_tider.append(newtons_metode(y, theta))
    kast_x_lengder.append(x(kast_tider[-1], theta))
    kontrolltider.append(36*np.sin(theta)/9.81)
```

**Lager plot av de ulike kastene**

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
i = 0

plt.figure(figsize=(14,5))
plt.title('Kast med ulik vinkel (radianer)')
plt.xlabel('$x, lengde (m)$') # Merker x-aksen
plt.ylabel('$y, høyde (m)$') # Merker y-aksen
for theta in theta_verdier:
    t_tider = np.arange(0, kast_tider[i], 0.001)
    plt.plot(x(t_tider, theta), y(t_tider, theta), label=f'$\Theta = \pi/{i+3}$') 
    i+=1

plt.legend()
plt.show()
```

#### a) 

```python
print(f'Ballen treffer bakken etter {kast_tider[1]:.2f} s. Kontroll: {kontrolltider[1]:.2f} s')
```

#### b)

```python
print(f'Ballen treffer bakken på {kast_x_lengder[1]:.2f} m')
```

#### c) 

Vi kan se at det lengste kastet skjer ved en vinkel på $\pi/4$ radianer.

```python
maks = max(kast_x_lengder)
indeks = kast_x_lengder.index(maks)
theta = theta_verdier[indeks]

print(f'Den lengste kastet på {maks:.2f} m skjer med vinkel på {theta:.3} radianer')
```

###    ( iii )  Sammenlign Newtons metode med Halveringsmetoden

Sammenlign Newtons metode med Halveringsmetoden for å løse oppgave a). Plot tilnærmingsverdiene mot hverandre for hvert steg fra 0 til 12.   


```python
def halveringsmetoden(y, a=0.5, b=1.5, N=12):
    h_list = []; 
    for n in range(N):
        m = (a + b)/2
        if y(a)*y(m) < 0:
            b = m
        else:
            a = m
        
        h_list.append(m);
    return m, h_list


def newtons_metode(y, t=2, N=12, h=10**-5):
    n_list = [];
    for n in range(N):
        t = t - y(t)/((y(t + h) - y(t))/h)    
        n_list.append(t)
    return t, n_list
```

```python
plt.figure(figsize=(14,6))
plt.title('Sammenligning av to ulike algoritmer for å finne nullpunkt')
plt.xlabel('N, antall steg') # Merker x-aksen
plt.ylabel('Nullpunkt') # Merker y-aksen
plt.plot(newtons_metode(y)[1], label='Newtons metode')
plt.plot(halveringsmetoden(y)[1], label='Halveringsmetoden')
plt.legend()
plt.show()
```

Vi kan se at Newtons metode går mye raskere mot nullpunktet. Tidsbruken er ganske lik.

```python
%timeit newtons_metode(y)
```

```python
%timeit halveringsmetoden(y)
```

$$f(x)=\frac{100x^2+1000}{1.07x^3+45}$$

```python
def f(x):
    return (100*x**2 + 1000)/(1.07*x**3+45)
```

```python
def derivert(f, a, h=10**-8):
    return (f(a + h) - f(a))/h
```

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'


x = np.linspace(0,100, 10**4)
fDer = derivert(f,x)

fDerTp = (fDer.argmax()/100, np.max(fDer))

plt.figure(figsize=(14,6))
plt.title('Dyrebestand etter x-år')
plt.xlabel('År')
plt.ylabel('Bestand')
plt.plot(x, f(x),  label='$f(x)$')
plt.plot(x, fDer, label='$f^\prime(x)$')
plt.plot(fDerTp[0], fDerTp[1], 'o', ms=5)
plt.annotate(f'({fDerTp[0]:.2f}, {fDerTp[1]:.2f})', xy=(fDerTp[0], fDerTp[1]), xytext=(-4, 7),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.axhline(0, color='gray', lw=0.5)
plt.legend()
plt.show()


print(f'Dyrebestanden vokser raskest ved {fDerTp[0]:.2f} år. Vekstfart på {fDerTp[1]:.2f}')
```

```python
import pandas as pd
url ='https://www.uio.no/studier/emner/matnat/math/MAT-INF1100/h18/obliger/running.txt'
df = pd.read_csv(url, sep=',', header=None)
df.columns = ['tid', 'fart']
```

### ( i ) Plot farten med hensyn på tiden

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
plt.figure(figsize=(14,6))
plt.title('Farten til objektet')
plt.xlabel('Tid (s)') # Merker x-aksen
plt.ylabel('Fart (m/S)') # Merker y-aksen
plt.plot(df.tid, df.fart)
plt.show()
```

### ( ii ) Lag et array med akselerasjonen og plot resultatet

Vi må derivere farten for å bestemme akselerasjonen.

```python
akselerasjon = np.zeros(len(df.tid))
for i in range(len(akselerasjon)-1):
    akselerasjon[i] = (df.fart[i+1]-df.fart[i])/(df.tid[i+1]-df.tid[i])
```

```python
plt.figure(figsize=(14,6))
plt.title('Akselerasjonen til objektet')
plt.xlabel('Tid (s)')
plt.ylabel('Akselerasjonen (m/s$^2$)')
plt.plot(df.tid, akselerasjon)
plt.show()
```

```python
akselerasjon2 = np.diff(df.fart)/np.diff(df.tid)
plt.figure(figsize=(14,6))
plt.title('Akselerasjonen til objektet')
plt.xlabel('Tid (s)')
plt.ylabel('Akselerasjonen (m/s$^2$)')
plt.plot(df.tid[:-1], akselerasjon2)
plt.show()
```

### ( iii ) Gi en algoritme for å beregne en tilnærming til objektets avstand $s(t)$ fra startpunktet ut fra de beregnede verdiene når $v(t) = s′(t)$ og $s(t_0) = 0$.

```python
distanse = np.zeros(len(df.tid))
distanse[0] = 0
for i in range(len(akselerasjon)-1):
    distanse[i+1] = distanse[i] + df.fart[i+1]*(df.tid[i+1]-df.tid[i])
```

```python
plt.figure(figsize=(14,6))
plt.title('Avstand fra startpunktet')
plt.xlabel('Tid (s)')
plt.ylabel('Distanse (m)')
plt.plot(df.tid, distanse)
plt.show()
```

## Oppgave 4
Utled og forklar Eulers metode. 
___

<img src='euler3.png' width='800px'>

“Neste $y$ er forrige $y$ pluss $h$ ganger endringsrate”. I rekkefølge beregnes $y_0$, $y_1$, $y_2$.....

Tenk deg at vi skal beregne formen på en ukjent kurve til $y(t)$ (den blå grafen over) som starter ved et gitt punkt og tilfredstiller differensialligningen  $y^\prime(t)=f(t, y(t))$. Her kan vi bruke differensialligningen som en formel for å beregne stigningtallet til tangentlinjen til kurven hvor som helst på kurven når posisjonen til det punktet er beregnet.

Tanken er at mens kurven i utgangspunktet er ukjent, er utgangspunktet $(t_0, y_0)$ kjent. Derfra kan vi bruke differensialligningen til å beregne en tilnærming til den momentane vekstfarten til y og stigningtallet til tangentlinjen.


Ta et lite steg langs tangentlinja til punktet $(t_1, y_1)$ I dette lille steget vil ikke vekstfarten til kurven endre seg så mye. Det betyr at $(t_1, y_1)$ vil være veldig nærme kurven vi ønsker å tilnærme. Slik gjentar vi prosessen. Velger vi små nok steg vil tilnærmingen være svært god.


Likning: $y^\prime(t)=f(t, y(t)), \; y(0)=y_0$

I punktet $x_k=k\cdot h$ tilnærmer vi 

$y^\prime = \frac{y_{k+1}-y_k}{h}$


og setter inn i likningen og løser den med hensyn på $y_{k+1}$

\begin{align}
\frac{y_{k+1}-y_k}{h} & = f(t_k, y_k)\\
y_{k+1}-y_k & = h\cdot f(t_k, y_k)\\
y_{k+1} & = y_k + h\cdot f(t_k, y_k)\\
\end{align}
___

## Oppgave 5
Løs diffligningene numerisk og plot $y(t),\; t\in[0,5]$:
___

```python
def diff(f, y_0, T, h = 2**(-7)):
    N = round(T/h) #Tid delt på steglengde
    t = np.linspace(0, T, N+1)
    y = np.zeros(N+1)

    # Fyller arrayet med funskjonsverdier
    y[0] = y_0 
    for k in range(N):
        y[k+1] = y[k] + h*f(y[k],k)

    # Plotter grafen til y
    plt.figure(figsize=(14,6))
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.plot(t, y)
    plt.show()
```

```python
def f(y_k, k):
    return k + y_k

diff(f, 0, 5, h = 2**(-7))
```
