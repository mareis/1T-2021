### Oppgave 1.1: Gausskurven

Den bjelleformede gausskurven (tetthetsfunksjonen til normalfordelingen),
$$ f(x)=\frac{1}{\sqrt{2\pi}\, s}e^{-\frac{1}{2}(\frac{x-m}{s})^2} $$
er en av de mest brukte funksjonene i vitenskap og teknologi. Parameterene $m$ og $s>0$ er reelle tall. Plot funksjonen, $f$, for $x \in [-10, 10]$. Hva skjer når du endrer på $m$ og $s$?

$\mu$ og $\sigma$ er erstattet med $m$ og $s$ i funksjonen.

<img src='http://www.muelaner.com/wp-content/uploads/2013/07/Standard_deviation_diagram.png'>

Funkejson er oppkalt etter Carl Friedrich Gauss, 1777-1855, som var et tysk matematiker og vitenskapsmann. Han er nå regnet som en av de mest betydningsfulle vitenskapsmennene gjennom tidene. Han bidro innen; tallteori, statistikk, matematisk analyse, differentialgeometri, astronomi og optikk. Gauss introduserte funksjonen når han analyserte sannsynligheter forbundet med astronomi-data. Les mere på [Wikipedia](https://no.wikipedia.org/wiki/Normalfordeling).

### Løsning:

```python
m = 0
s = 2
x = np.linspace(-10, 10, 1000)
y = 1/(np.sqrt(2*np.pi)*s) * np.e**(-(1/2)*((x - m)/s)**2)
plt.plot(x, y)
plt.title('Normalfordelingen')
plt.show()
```
