## Grunnleggende om matplotlib

Matplotlib er et bibliotek, og må importeres før det kan brukes. 

Importerer biblioteket og setter formatet til svg

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

Listene over danner punktene (0,0), (1,1), (2,4), (3,9) og (4,16) som hver og en blir plottet og forbundet med en rett linje.

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
```

Lage plottet

```python
plt.plot(x, y)
plt.show()
```

Som regel ønsker vi mer informasjon i plottet. I eksempelet under ser vi hvordan vi kangjøre det.

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'

plt.xlabel('$x$')      # Merker x-aksen
plt.ylabel('$f(x)$')   # Merker y-aksen
plt.title('En tittel') #Tittel
plt.plot(x, y, '-.', label='$f(x)=x^2$')
plt.axhline(0, color='gray', lw=0.5) 
plt.legend()           # Infoboksen øverst i hjørnet. Bruker data fra label
plt.show()
```

## Numpy og funksjoner i matplotlib

$$f(x) = x^2 -3x +7$$

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'

def f(x):
    return x**2 - 3*x + 7

x = np.linspace(-1, 4, 100) # Lager 100 x-verdier fra -1 til til og med 4

plt.plot(x, f(x))         # Plotter x og y-verdiene
plt.xlabel('x')           # Merker x-aksen
plt.ylabel('$f(x)$')      # Merker y-aksen
plt.show()
```

### Egndefinerte funksjoner

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'

def f(t):
    return t**2 + 3*t + 1

t = np.linspace(-4, 1, 100)

plt.plot(t, f(t), label='$f(t)=x^2+2x+1$')
plt.xlabel('t')           # Merker x-aksen
plt.ylabel('$f(t)$')      # Merker y-aksen
plt.axhline(0, color='gray', lw=0.5)
plt.legend()
plt.show()
```

### To funksjoner i samme figur

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'

def f(x):
    return x**2

def g(x):
    return -x**2 + 10

x = np.linspace(-5, 5, 100)

plt.plot(x, f(x), label='$f(x)=x^2$')
plt.plot(x, g(x), label='$g(x)=-x^2+10$')
plt.legend()
plt.show()
```
