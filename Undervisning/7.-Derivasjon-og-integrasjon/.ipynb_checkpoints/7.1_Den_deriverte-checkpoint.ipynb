# 7.1 Derivasjon

### Litt om rette linjer.
En rett linje har likningen $y=ax+b$, der a er stigningstallet.
Vi kan definere stigningstallet til en rett linje dersom vi vet koordinatene til to punkter på linja. Da har vi 

$$a=\frac{y_2−y_1}{x_2−x_1},$$

hvor $(x_1,y_1)$ er koordinatet til det første punktet, og $(x_2,y_2)$ er koordinatet til det andre punktet. Denne formelen kalles ofte for "topunktsformelen"

Når vi ønsker å vite stigningen til en funksjon/graf i et bestemt punkt, så trenger vi å vite stigningstallet til den rette linjen som såvidt er borti (tangerer) funksjonen/grafen i dette punktet. En slik linje kaller vi en tangent og definerer den som en linje som deler kun ett punkt med funksjonen. Rett ovenfor så vi derimot at vi trenger to punkter for å kunne regne ut stigningstallet til en rett linje. Vi må derfor ha en linje som går gjennom to punkter på funksjonen/grafen; vi kaller en slik linje for en sekant. La oss oppsummere så langt:


### Tangent og sekant.
* En tangent er en linje som snitter eller er såvidt borti en graf. Rent matematisk sier vi at den deler kun ett punkt med grafen.
* En sekant er en linje som skjærer grafen i to punkter. Vi kan bruke en slik linje til å regne ut stigningstallet til linjen med topunktsformelen.


#### Numerisk derivasjon
Numerisk derivasjon går tilbake til den grunnleggende idéen - stigningstall. Den enkleste numeriske metoden er rett og slett å begynne med definisjonen av den deriverte i et punkt x. Vi endrer notasjonen litt og setter $\Delta x = x_2 - x_1$, $f(x+\Delta x)=y_2$ og $f(x)=y_1$.



$$f^\prime(x)=\lim_{\Delta x \to 0} \frac{f(x+\Delta x) - f(x)}{\Delta x}$$


velge  $\Delta x$ som et lite tall og regne ut

$$f^\prime(x) \approx \frac{f(x + \Delta x) - f(x)}{\Delta x}$$

## Newtons kvotient
Denne metoden kalles Newtons kvotient og er en tilnærming til den deriverte. Den deriverte til en funksjon $f$ kan tilnærmes slik dersom $h$ er liten.

$$f^\prime(a) \approx \frac{f(a + h) - f(a)}{h}$$


## Implimentering i Python

```python
def derivert(f, a, h=10**-8):
    return (f(a + h) - f(a))/h
```

### Eksempel 1

```python
def f(x):
    return x**2

print(f'{derivert(f,2):.7f}')
```

### Begrensninger med å derivere på en datamaskin.
* h kan ikke ha verdien 0 ettersom det å dele på 0 er matematisk udefinert.
* Å jobbe med ekstremt små tall på en PC gir ofte unøyaktige svar.
* Hvis vi bruker store verdier for h blir svarene også unøyaktige.

### Eksempel 2

$$g(x)=x^4 - 1$$

$$g^\prime (3)= 4 \cdot 3^3=108$$

Plot diferansen mellom den eksakt deriverte til $g^\prime (3)$ og den tilnærmede numeriske med ulike $h$.


```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'

def g(x):
    return x**4 - 1

avvikene = []
h = []
for i in range(3, 14):
    deriverte = derivert(g, 3, 10**-i)
    avvik = np.abs(deriverte - 108)
    avvikene.append(avvik)
    h.append(f'10e-{i}')

plt.figure(figsize=(10,4))
plt.xlabel('h') 
plt.ylabel('$g^{\prime}(3)$')
plt.title('Analytisk vs numerisk ved ulike h') 
plt.plot(h, avvikene)
plt.show()
```
